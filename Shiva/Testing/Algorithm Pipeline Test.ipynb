```python
class Agent:
    def __init__(self, input_dim, output_dim):
        self.policy = torch.nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, output_dim),
        )
        self.target_policy = torch.nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, output_dim),
        )
        self.optimizer = torch.optim.Adam(params=self.policy.parameters(), lr=0.001)
```

```python
env.get_observation_space(), env.get_action_space()
```

```python
agent = Agent(env.get_observation_space()+env.get_action_space(), 1)
```

```python
env.reset()
```

```python
num_of_episodes = 50

for i in range(num_of_episodes):
    obs = env.reset()
    obs = env.get_observation()
    done = False
    while not done:
        observation = env.get_observation()
        
        action = alg.get_action(agent, observation, env.get_current_step())

        next_observation, reward, done = env.step(action)
        expr = [observation, action, reward, next_observation, int(done)]
        buffer.append(expr)
        
        sample = buffer.sample()
        
        alg.update(agent, sample, env.get_current_step())
        
        observation = next_observation
```
